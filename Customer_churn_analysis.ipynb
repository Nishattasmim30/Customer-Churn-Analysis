```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


#For better visuals
sns.set(style="whitegrid")
```

```python
url = "https://raw.githubusercontent.com/0Theo/Bank-Customer-Churn-Dataset/main/Bank%20Customer%20Churn%20Prediction.csv"
df = pd.read_csv(url)
df.head()
```

```python
df.info()
df.isnull().sum()
```

```python
df = df.drop(['customer_id', 'unused variable'], axis=1, errors='ignore')
df.head()
```

```python
sns.countplot(x='churn', data=df)
plt.title("Churn: 1 = Left, 0 = Stayed")
plt.show()
```

```python
sns.boxplot(x='churn', y='age', data=df)
plt.title("Age by Churn")
plt.show()
```

```python
sns.boxplot(x='churn', y='balance', data=df)
plt.title("Balance by Churn")
plt.show()
```

```python
sns.countplot(x='country', hue='churn', data=df)
plt.title("Churn by Country")
plt.show()

sns.countplot(x='gender', hue='churn', data=df)
plt.title("Churn by Gender")
plt.show()
```

```python
df_encoded = pd.get_dummies(df, drop_first=True)
df_encoded.head()
```

```python
corr = df_encoded.corr()
corr['churn'].sort_values(ascending=False)
```
